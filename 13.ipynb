```python
#carga de librerias
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import sklearn.metrics
import sklearn.neighbors
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OrdinalEncoder
import sklearn.preprocessing

import itertools 

import math
import sklearn.metrics

from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score

from IPython.display import display

import random
import time
```

```python
#Carga y exploracionde los datos
df=pd.read_csv('/datasets/car_data.csv')
display(df)
df.info()
```

## Análisis del modelo

```python
#mrse = mean_squared_error(target_test, y_pred)
#r2 = r2_score(target_test, y_pred)

#print(f"Error cuadrático medio (MRSE): {mrse:.2f}")
#print(f"Coeficiente de determinación (R²): {r2:.2f}")

def mrse_r2(real, pred):
    mrse=mean_squared_error(real, pred)
    r2=r2_score(real, pred)
    print(f"Error cuadrático medio (MRSE): {mrse:.2f}")
    print(f"Coeficiente de determinación (R²): {r2:.2f}")
```

Modelo de Regresion lineal con codificacion ordinal

```python
start_time = time.time()

model_lr = LinearRegression()
model_lr.fit(features_train, target_train)

training_time = time.time() - start_time
print(f"⏳ Tiempo de entrenamiento: {training_time:.2f} segundos")


start_time = time.time()
train_pred_lr = model_lr.predict(features_train)
valid_pred_lr = model_lr.predict(features_valid)
test_pred_lr = model_lr.predict(features_test)


print(f'RMSE entrenamiento{mrse_r2(target_train, train_pred_lr)}')
print(f'RMSE validacion{mrse_r2(target_valid, valid_pred_lr)}')
print(f'RMSE prueba{mrse_r2(target_test, test_pred_lr)}')
prediction_time = time.time() - start_time
print(f"⏳ Tiempo de predicción: {prediction_time:.4f} segundos")
```

usando codificacion One-Hot Encoding (OHE) 

```python
start_time = time.time()
model_lr = LinearRegression()
model_lr.fit(features_train_ohe, target_train_ohe)
training_time = time.time() - start_time
print(f"⏳ Tiempo de entrenamiento: {training_time:.2f} segundos")


start_time = time.time()
train_pred_lr_ohe = model_lr.predict(features_train_ohe)
valid_pred_lr_ohe = model_lr.predict(features_valid_ohe)
test_pred_lr_ohe = model_lr.predict(features_test_ohe)

print(f'RMSE entrenamiento{mrse_r2(target_train_ohe, train_pred_lr_ohe)}')
print(f'RMSE validacion{mrse_r2(target_valid_ohe, valid_pred_lr_ohe)}')
print(f'RMSE prueba{mrse_r2(target_test_ohe, test_pred_lr_ohe)}')
prediction_time = time.time() - start_time
print(f"⏳ Tiempo de predicción: {prediction_time:.4f} segundos")
```

```python
from catboost import CatBoostRegressor

model_cbc = CatBoostRegressor(
    iterations=500,         # Número de iteraciones (árboles)
    depth=8,                # Profundidad del árbol
    learning_rate=0.05,     # Tasa de aprendizaje
    loss_function='RMSE',   # Función de pérdida para regresión

    verbose=50,             # Mostrar cada 50 iteraciones
    random_seed=12345)

start_time = time.time()
model_cbc.fit(features_train, target_train, eval_set=(features_test, target_test))
training_time = time.time() - start_time
print(f"⏳ Tiempo de entrenamiento: {training_time:.2f} segundos")


start_time = time.time()
train_pred_cbc = model_cbc.predict(features_train)
valid_pred_cbc = model_cbc.predict(features_valid)
test_pred_cbc = model_cbc.predict(features_test)

print(f'RMSE entrenamiento{mrse_r2(target_train, train_pred_cbc)}')
print(f'RMSE validacion{mrse_r2(target_valid, valid_pred_cbc)}')
print(f'RMSE prueba{mrse_r2(target_test, test_pred_cbc)}')
prediction_time = time.time() - start_time
print(f"⏳ Tiempo de predicción: {prediction_time:.4f} segundos")
```

```python
import lightgbm as lgb
# Create an instance of the LightGBM Regressor with the RMSE metric.
model_lgb = lgb.LGBMRegressor(num_interaction=1000, verbose=1, metric='rmse')

# Train the model using the training data.
start_time = time.time()
model_lgb.fit(features_train, target_train, eval_set=(features_train, target_train))
training_time = time.time() - start_time
print(f"⏳ Tiempo de entrenamiento: {training_time:.2f} segundos")

# Make predictions on the training and validation data.
start_time = time.time()
train_pred_lgb = model_lgb.predict(features_train)
valid_pred_lgb = model_lgb.predict(features_valid)
test_pred_lgb = model_lgb.predict(features_test)

print(f'RMSE entrenamiento{mrse_r2(target_train, train_pred_lgb)}')
print(f'RMSE validacion{mrse_r2(target_valid, valid_pred_lgb)}')
print(f'RMSE prueba{mrse_r2(target_test, test_pred_lgb)}')
prediction_time = time.time() - start_time
print(f"⏳ Tiempo de predicción: {prediction_time:.4f} segundos")
```

```python
import xgboost as xgb

model_xgb = xgb.XGBRegressor(
    objective='reg:squarederror',  # Función de pérdida para regresión
    n_estimators=100,
    learning_rate=0.05,
    max_depth=10,
    random_state=12345
)
start_time = time.time()
model_xgb.fit(features_train_ohe, target_train_ohe)
training_time = time.time() - start_time
print(f"⏳ Tiempo de entrenamiento: {training_time:.2f} segundos")


start_time = time.time()
train_pred_xgb = model_xgb.predict(features_train_ohe)
valid_pred_xgb = model_xgb.predict(features_valid_ohe)
test_pred_xgb = model_xgb.predict(features_test_ohe)

print(f'RMSE entrenamiento{mrse_r2(target_train_ohe, train_pred_xgb)}')
print(f'RMSE validacion{mrse_r2(target_valid_ohe, valid_pred_xgb)}')
print(f'RMSE prueba{mrse_r2(target_test_ohe, test_pred_xgb)}')
prediction_time = time.time() - start_time
print(f"⏳ Tiempo de predicción: {prediction_time:.4f} segundos")
```
